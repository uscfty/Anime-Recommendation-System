```python
import pandas as pd

file_path = "/Users/futianyu/Untitled Folder 1/anime.csv"  
data = pd.read_csv(file_path)

# Display the first few rows of data
print(data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the chart style
sns.set(style="whitegrid")

# Rating distribution histogram
plt.figure(figsize=(10, 6))
sns.histplot(data['rating'], bins=20, kde=True, color='blue')
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Anime Rating Distribution")
plt.show()
```

```python
# Count the number of animations by type
plt.figure(figsize=(10, 8))
data['type'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel("Anime Type")
plt.ylabel("Count")
plt.title("Anime Type Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of ratings and number of members
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['members'], y=data['rating'], alpha=0.5, color='red')
plt.xscale("log")  # Use a logarithmic scale to better show the data distribution
plt.xlabel("Members (log scale)")
plt.ylabel("Rating")
plt.title("Anime Rating vs Members")
plt.show()
```

```python
# Average ratings of different types of animation
genre_ratings = data.groupby('type')['rating'].mean().sort_values()
plt.figure(figsize=(10, 6))
genre_ratings.plot(kind='barh', color='green', edgecolor='black')
plt.xlabel("Average Rating")
plt.ylabel("Anime Type")
plt.title("Average Rating by Anime Type")
plt.show()
```
